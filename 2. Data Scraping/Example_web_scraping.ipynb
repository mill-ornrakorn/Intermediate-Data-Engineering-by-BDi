## POST Method

```python

#Prep payload from json string
# payload_str =  "{}"
# payload = json.loads(payload_str)

#Prep payload
payload = {
    "title": "foo",
    "body": "bar",
    "userId": 1,
  }

#Perp header
header = {
"user-agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/105.0.0.0 Safari/537.36"
}
```
